```python
# count of distinct order
orders['order_id'].nunique()
```

```python
# distinct restaurant name
orders['restaurant_name'].unique()
```

```python
# restaurant with most order rank
orders.groupby('restaurant_name')['order_id'].nunique()
```

```python
orders_new = orders.groupby('restaurant_name')['order_id'].nunique().reset_index()
orders_new.sort_values(by=['order_id'],ascending = False)
```

```python
# rank
orders_new['Rank']=orders_new['order_id'].rank(method='first', ascending = False)
```

```python
# top rank
orders_new[orders_new['Rank']==1]
```

```python
# change of revenue year to year
orders.info()
```

```python
orders['order_time'] = orders['order_time'].apply(pd.to_datetime)
orders.info()
```

```python
orders['Year'] = orders['order_time'].dt.year
orders
```

```python
# total revenue of years
orders2 = orders.groupby('Year')['order_total'].sum().reset_index()
orders2
```

```python
# comparing to previous year
orders2['previousYearSale']=orders2['order_total'].shift(1)
orders2
```

```python
orders2['%diff']= (orders2['order_total']-orders2['previousYearSale'])/orders2['previousYearSale']
orders2
```
